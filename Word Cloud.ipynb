```python
from wordcloud import WordCloud
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cdf=pd.read_csv("C:/Users/rabhi/OneDrive/Documents/Completed_Trials.csv")
```

```python
all_words = cdf['Title']
wc = WordCloud(max_words = 1000 , width = 1600 , height = 800,
              collocations=False).generate(" ".join(all_words))
plt.figure(figsize = (20,20))
plt.imshow(wc)
all_words.shape
```

```python
all_words = cdf['Conditions']
wc = WordCloud(max_words = 1000 , width = 1600 , height = 800,
              collocations=False).generate(" ".join(all_words))
plt.figure(figsize = (20,20))
plt.imshow(wc)
all_words.shape
```
